```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ruta de la imagen
path = 'arte.jpeg'
img_bgr2 = cv2.imread("./arte.jpeg")
img_rgb = cv2.cvtColor(img_bgr2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
# CONVERTIR A ESCALA DE GRISES
# Dimensiones de la imagen original
dim = img_rgb.shape
# Nuevo arreglo bidimensional
img_gray = np.zeros((dim[0], dim[1]))
# Valor de intensidad con fraccion
frac = 1/3
for i in range(dim[0]):
    for j in range(dim[1]):
        new_r = img_rgb[i,j,0] * frac
        new_g = img_rgb[i,j,1] * frac
        new_b = img_rgb[i,j,2] * frac
        new_val = new_r + new_g + new_b
        img_gray[i,j] = np.ceil(new_val)
```

```python
plt.imshow(img_gray, 'gray')
plt.axis('off')
plt.show()
```
