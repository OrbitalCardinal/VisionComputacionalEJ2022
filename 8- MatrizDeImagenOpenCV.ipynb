## Cargar opencv y librerías para visualización de datos

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Cargar imagen con opencv

```python
# Funcion imread de opencv carga una imagen en forma de matriz por defecto
img = cv2.cvtColor(cv2.imread("./arte.jpeg"), cv2.COLOR_BGR2RGB)
```

## Mostrar imagen con matplotlib

```python
# plt es la libreria de visualización de datos, imshow es la función para mostrar matrices como si fueran imagenes
plt.imshow(img)
plt.axis('off')
plt.show()
```

## Imprimir matriz de valores de la imagen

```python
# La imagenes a color tienen 3 canales RGB, por lo que es un arreglo de tres dimensiones
img
```

```python
# Con la funcion shape se puede ver las dimensiones del arreglo (matriz) de la imagen
img.shape
```

```python
# Tambien se pueden eliminar los tres canales de colores si se le pasa el argumento 0, a la funcion imread
# Esto hace que se lea la imagen en un solo canal (escala de grises)
img_gray = cv2.imread("./flor.png",0)
```

```python
# Entonces ya no existe una dimension extra en el arreglo, solo dos dimensiones, cada valor en la matriz es
# la intensidad del pixel 0 mas negro a 255 mas blanco
img_gray.shape
```

```python
plt.imshow(img_gray,"gray")
```
